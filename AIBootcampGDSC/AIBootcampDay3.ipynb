# **Matplotlib**

```python
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 8, 15, 30])
y = np.array([0, 45 ,150, 200])
```

```python
plt.plot(x, y, marker="o")
plt.show()
```

```python
x2=np.array([5, 15, 24, 35])
y2=np.array([110, 80, 45, 5])
plt.plot(x,y)
plt.plot(x2,y2)
plt.show()
```

```python
plt.plot(x,y)
plt.plot(x2,y2)
plt.title("baslik")
plt.xlabel("x ekseni")
plt.ylabel("y ekseni")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)

plt.subplot(1,2,2)
plt.plot(x2,y2)
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
np.random.seed(19680801) 
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

fig, ax = plt.subplots(figsize=(8, 5))
ax.scatter('a', 'b', c='c', s='d', data=data)
ax.set_xlabel('entry a')
ax.set_ylabel('entry b');
```

```python
x = np.linspace(0, 2, 100) 


fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(x, x, label='lineer')  
ax.plot(x, x**2, label='karesel')  
ax.plot(x, x**3, label='kubik') 
ax.set_xlabel('x ekseni')  
ax.set_ylabel('y ekseni')  
ax.set_title("baslik")
ax.legend(); 
```

# **Seaborn**

```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

## **Kategorik verilerin görselleştirilmesi**

```python
df["sex"].value_counts()
```

```python
sns.countplot(x="sex",data=df)
plt.show()
```

```python
sns.catplot(data=df, kind="bar", x="day", y="total_bill", hue="smoker") #split True [violin, bar, swarm]
plt.show()
```

## **Numerik verilerin görseleştirilmesi**

```python
sns.lineplot(y=df["tip"],x=df["size"],data=df)
plt.show()
```

```python
sns.boxplot(x=df["total_bill"])
plt.show()
```

```python
df["total_bill"].hist()
plt.show()
```

```python
sns.displot(data=df, x="total_bill", col="time", kde=True)
plt.show()
```

```python
sns.relplot(
    data=df,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
plt.show()
```

```python
sns.jointplot(data=df, x="tip", y="total_bill",hue="sex")
plt.show()
```
